```python
# daily return
#daily price change
# open to close difference
```

```python
one2['Daily_Return']=one2['Close'].pct_change()
one2['Price_change']=one2['Close'].diff()
one2['Open_close_diff']=one2['Close']-one2['Open']
```

```python
one3=one2
```

```python
one3=one3.drop(['Daily_Return','Price_change','Open_close_diff'],axis=1)
```

```python
one3
```

```python
#moving averages and Exponential Moving averages /RSI
```

```python
one3['SMA_5'] = one3['Close'].rolling(5).mean()
one3['SMA_20'] = one3['Close'].rolling(20).mean()
one3['SMA_50'] = one3['Close'].rolling(50).mean()
one3['10D_EMA']=one3['Close'].rolling(window=10).mean()
one3['50D_EMA']=one3['Close'].rolling(window=50).mean()
change = one3['Close'].diff()
gain = change.clip(lower=0)
loss = -change.clip(upper=0)

avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()

rs = avg_gain / avg_loss
one3['RSI_14'] = 100 - (100 / (1 + rs))
```

```python
one3
```

```python
# Bollinger Band
```

```python
one3['SMA_20'] = one3['Close'].rolling(20).mean()
sd = one3['Close'].rolling(20).std()

one3['Upper_band'] = one3['SMA_20'] + (2 * sd)
one3['Lower_band'] = one3['SMA_20'] - (2 * sd)

```

```python
one3
```

```python
latest=one3
```

```python
latest.plot(x='Date',y=['Upper_band','Lower_band'],figsize=(10,4))
```
